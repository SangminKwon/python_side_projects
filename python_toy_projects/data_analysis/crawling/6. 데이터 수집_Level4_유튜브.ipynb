```python
from selenium import webdriver as wd
import time

driver = wd.Chrome('./chromedriver.exe')
target_site = 'https://www.youtube.com/results?search_query=%ED%9D%91%EC%9A%B4%EC%9E%A5'
driver.get(target_site)
# ëŒ€ê¸°í•˜ëŠ” ì½”ë“œëŠ” ê°•ì œì (time ëª¨ë“ˆ í™œìš©), ì•”ë¬µì (selenium ë¼ì´ë¸ŒëŸ¬ë¦¬ í™œìš©) ë“±ë“± ì¡´ì¬
time.sleep(2)
```

```python
# ê²Œì‹œë¬¼ì„ í•„ìš”í•œ ë§Œí¼ ê°€ì ¸ì˜¤ê¸° ìœ„í•´ì„œ, ìŠ¤í¬ë¡¤ë°”ë¥¼ ë‚´ë¦°ë‹¤
# 10íšŒ ì •ë„ ëê¹Œì§€ ë‚´ë ¤ì„œ ì¶”ê°€ ì˜ìƒ ëª©ë¡ì„ ë°›ì•„ì˜¨ë‹¤
# ìë°”ìŠ¤í¬ë¦½íŠ¸ë¥¼ íŒŒì´ì¬ì—ì„œ ëª…ë ¹ì„ ë‚´ë ¤ì„œ êµ¬ë™ì‹œì¼œì•¼ í•œë‹¤
for n in range(10):
    driver.execute_script('''
        window.scrollBy(0, 1000)
    ''')
    time.sleep(2)
```

```python
# ì˜ìƒë¬¼ì˜ ì œëª© ì¶”ì¶œ
# ì œëª©ì„ íŠ¹ì •í•´ì„œ ì¶”ì¶œ
import os 
import urllib.request as rq

videos = driver.find_elements_by_id('video-title')
counts = driver.find_elements_by_css_selector('div#metadata-line > span:nth-of-type(1)')
times = driver.find_elements_by_css_selector('div#metadata-line > span:nth-of-type(2)')

# ì œëª©, ì¡°íšŒìˆ˜, ì—…ë¡œë“œ ì‹œê°„ ë™ì‹œ í‘œê¸°

for video, count, time in zip(videos, counts, times):
    print(f'{video.text}  |  {count.text}  |  {time.text}')
    print()


thumbnails = driver.find_elements_by_css_selector('a#thumbnail > yt-img-shadow > img')
img_url =[]

for thumbnail in thumbnails:
    url = thumbnail.get_attribute('src')
    img_url.append(url)



img_folder = './img'

if not os.path.isdir(img_folder) : # ì—†ìœ¼ë©´ ìƒˆë¡œ ìƒì„±í•˜ëŠ” ì¡°ê±´ë¬¸ 
    os.mkdir(img_folder)
    
for index, link in enumerate(img_url) :
#     start = link.rfind('.')
#     end = link.rfind('&')
#     filetype = link[start:end]
    rq.urlretrieve(link, f'./img/{index}.jpg')
```
